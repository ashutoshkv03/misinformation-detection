```python
import pandas as pd

df = pd.read_csv("../data/raw/pheme_dataset.csv")
df.head()
```

```python
df.info()
```

```python
df["is_rumor"].value_counts()
```

```python
df["text_length"] = df["text"].astype(str).apply(len)
df[["text", "text_length", "is_rumor"]].head()
```

```python
import matplotlib.pyplot as plt

df["text_length"].hist(bins=40)
plt.title("Text Length Distribution")
plt.xlabel("Text Length")
plt.ylabel("Frequency")
plt.show()
```

```python
df.groupby("topic")["is_rumor"].mean().sort_values(ascending=False)
```
